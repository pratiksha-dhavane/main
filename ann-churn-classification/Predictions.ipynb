```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
import pickle

import tensorflow as tf
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
import datetime
```

```python
# Load the trained model, scaler pickle file

model = load_model(r"/content/drive/MyDrive/DeepLearning/Churn_Modelling/model_v1.h5")

with open(r"/content/drive/MyDrive/DeepLearning/Churn_Modelling/label_encoder_gender.pkl","rb") as file:
  label_encoder_gn = pickle.load(file)

with open("/content/drive/MyDrive/DeepLearning/Churn_Modelling/label_encoder_geo.pkl","rb") as file:
  label_encoder_geo = pickle.load(file)

with open("/content/drive/MyDrive/DeepLearning/Churn_Modelling/scaler.pkl","rb") as file:
  scaler= pickle.load(file)
```

```python
# Example Input Data

input_data = {
    'CreditScore' : 600,
    'Geography' : 'France',
    'Gender' : 'Male',
    'Age' : 40,
    'Tenure' : 3,
    'Balance' : 60000,
    'NumOfProducts' : 2,
    'HasCrCard' : 1,
    'IsActiveMember' : 1,
    'EstimatedSalary' : 50000
}

input_data = pd.DataFrame([input_data])
```

```python
geo_encoder = label_encoder_geo.transform(input_data[['Geography']])
get_encoder_df = pd.DataFrame(geo_encoder,columns=label_encoder_geo.get_feature_names_out())
get_encoder_df
```

```python
input_data['Gender'] = label_encoder_gn.transform(input_data['Gender'])
```

```python
input_df = pd.concat([input_data.drop('Geography',axis=1),get_encoder_df],axis=1)
```

```python
input_scaled = scaler.transform(input_df)
```

```python
prediction = model.predict(input_scaled)
prediction
```

```python
prediction_proba = prediction[0][0]
```

```python
if prediction_proba > 0.5:
  print("Customer is likely to Churn.")
else:
  print("Customer is not likely to Churn.")
```
